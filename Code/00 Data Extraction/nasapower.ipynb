```python
import yaml
from pathlib import Path
import pandas as pd
from src.power_api import PowerAPI
from datetime import datetime
```

```python
from power_api import PowerAPI
```

```python
def main():
    output_dir = "./data"
    cities_file = 'cities.csv'
    cities_df = pd.read_csv(cities_file)

    start_date = '20160101'     #YYYYMMDD
    end_date = '20211231'       #YYYYMMDD

    for index, row in cities_df.iterrows():
        city_name = row['city']
        latitude = row['lat']
        longitude = row['lon']

        output_file = Path(output_dir) / f"{city_name}.csv"

        print(f"{datetime.now()} - Downloading data for {city_name}...")

        nasa_weather = PowerAPI(start=pd.Timestamp(str(start_date)), end=pd.Timestamp(str(end_date)), long=longitude, lat=latitude)

        weather_df = nasa_weather.get_weather()

        weather_df.to_csv(output_file, sep=";")

        print(f"{datetime.now()} - Data for {city_name} successfully downloaded and saved to {output_file}")
```

```python
if __name__ == '__main__':

    main()
```
